Problem : Maximum sum of subArray of size K

 I/P = array , size, k 
2	5	1	8	2	9	1

Window Size k = 3 

Find maximum sum for subArray with size k 

[Maximum sum of subArray of size K](https://www.geeksforgeeks.org/problems/max-sum-subarray-of-size-k5313/1)



```python
class Solution:
    def maxSubarraySum(self, arr, k):
        # code here 
        i,j,max_sum,sum=0,0,0,0
        n = len(arr)
        while(j<n):
            sum += arr[j]
            if j-i+1 <k:
                j+=1
            else:
                max_sum = max(max_sum,sum)
                sum-=arr[i]
                i+=1
                j+=1
        return max_sum
```
