Given an array arr[]  and a positive integer k, find the first negative integer for each and every window(contiguous subarray) of size k.

Note: If a window does not contain a negative integer, then return 0 for that window.

Examples:

Input: arr[] = [-8, 2, 3, -6, 10] , k = 2
Output: [-8, 0, -6, -6]
Explanation:
Window [-8, 2] First negative integer is -8.
Window [2, 3] No negative integers, output is 0.
Window [3, -6] First negative integer is -6.
Window [-6, 10] First negative integer is -6.


Link: [First negative in every window of size k](https://www.geeksforgeeks.org/problems/first-negative-integer-in-every-window-of-size-k3345/1)

```python
from typing import List
from collections import deque
```

```python
class Solution:
    def first_negative_integer_for_subarray(self,arr:List, window_size:int)->List: 
        if arr ==[] or window_size ==0:
            return []
        else :
            neg_integer = deque()
            output_arr = []
            i,j = 0,0
            n = len(arr)
            while(j<n):
                if arr[j]<0:
                    neg_integer.append(arr[j])
                if (j-i +1) < window_size:
                    j+=1
                    # if arr[j]<0:
                    #     neg_integer.append(arr[j])
                else:
                    if len(neg_integer)>0:
                        output_arr.append(neg_integer[0] )
                        print(arr[i], neg_integer[0], arr[i] == int(neg_integer[0]),j)
                        if arr[i] == int(neg_integer[0]):
                            neg_integer.popleft()
                        print(neg_integer)                   
                            
                    else:
                        output_arr.append(0)
                    i+=1
                    j+=1
                    
            return output_arr
```

```python
a = Solution()
print(a.first_negative_integer_for_subarray([-8, 2, -3, -6, 10,11,-2,9,0], 5))
```
