#### Question:  https://leetcode.com/problems/remove-duplicates-from-sorted-array-ii/description/ ####


Ideation to solve : 
1. Two pointer approach. 
2. First pointer walks through the array and second to replace
3. We will do replacement by checking count is more than 1.


```python
class Solution(object):
    def removeDuplicates(self, nums):
        """
        :type nums: List[int]
        :rtype: int
        """
        insert_pos = 1
        counter =0 
        for i in range(1, len(nums)-1):
            if nums[i] != nums[i + 1]:
                counter = 0
                nums[insert_pos] = nums[i]
                insert_pos += 1
            else: 
                counter += 1
                if counter < 2:
                    insert_pos += 1

        return insert_pos
```

```python
solution = Solution()
nums = [0,0,1,1,1,1,2,3,3]
print(nums[:solution.removeDuplicates(nums)+1])
```

## 2. What was Right vs. What was Wrong

### ✅ What went Right
* **Logic:** You correctly identified the core logic: compare `nums[i]` with `nums[i+1]`.
* **Variable Naming:** `first_idx` and `len_nums` are clear and descriptive.
* **Edge Case Handling:** You handled the shrinking size of the array correctly by decrementing `len_nums`.

### ❌ What went Wrong (Critical)
* **The `pop()` Trap:** You used `nums.pop(first_idx)` inside a `while` loop.
    * In Python, removing an item from the middle of a list requires shifting all subsequent elements to the left to fill the gap.
    * This operation costs **$O(N)$**.
* **Compounding Complexity:** Since you do this inside a loop that iterates $N$ times, your total time complexity becomes **$O(N^2)$**.
    * *Scenario:* If the input array has 100,000 elements (all duplicates), your solution performs approx. $100,000 \times 100,000$ operations. The interviewer expects this to run in $O(N)$.

---

## 3. Best Way to Implement: The Two-Pointer Approach

**Concept:**
Since the array is sorted, we don't need to physically "remove" items (which is expensive). We only need to **overwrite** the front of the array with the unique elements.

We use two pointers:
1.  **`insert_pos` (Slow Runner):** Points to the location where the *next* unique number should be written.
2.  **`i` (Fast Runner):** Scans ahead to find new unique numbers.

**Algorithm:**
1.  Start `insert_pos` at index 1 (since the first element is always unique).
2.  Iterate `i` from 1 to the end.
3.  If `nums[i]` is different from `nums[i-1]`, it means we found a new unique number.
    * Copy it to `nums[insert_pos]`.
    * Increment `insert_pos`.
4.  If they are the same, do nothing (skip the duplicate).
5.  Return `insert_pos` (which represents the count of unique elements).


       

```python
## Revised soltion with better performance

class Solution(object):
    def removeDuplicates(self, nums):
        """
        :type nums: List[int]
        :rtype: int
        """
        nums_len = len(nums)
        insert_pos = 1
        for i in range(1,nums_len):
            if nums[i] != nums[i-1]:
                nums[insert_pos] = nums[i]
                insert_pos +=1
        return insert_pos
```
